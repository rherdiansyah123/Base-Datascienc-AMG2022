```python
import matplotlib.pyplot as plt
```

```python
print("======== SOAL NO 1 ========")
#A-XIS
a = 300000
b = 20
c = 75000

hasil_a = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan A-Xis : " + str (hasil_a))


#B-LOG
a = 360000
b = 14
c = 75000

hasil_b = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan B-Log : " + str (hasil_b))


#C-in
a = 475000
b = 28
c = 75000

hasil_c = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan C-in : " + str (hasil_c))


#D-namic
a = 580000
b = 42
c = 75000

hasil_d = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan D-namic : " + str (hasil_d))


#E-psilon
a = 400000
b = 18
c = 75000

hasil_e = (a * b) + (c * b)
print("pengeluaran gaji + asuransi setiap perusahaan E-psilon : " + str (hasil_e))
```

```python
print("======== SOAL NO 2 ========")

pajak = 0.03
utang_a = 0.08
utang_b = 0.04
utang_c = 0.12
utang_d = 0.11
utang_e = 0.03

laba_a = 32500000
laba_b = 31000000
laba_c = 42000000
laba_d = 51500000
laba_e = 27000000

hasil_a = (pajak * laba_a) + (utang_a * 280000000)
print("maka total pajak dan utang A adalah : " + str(hasil_a))
hasil_b = (pajak * laba_b) + (utang_b * 310000000)
print("maka total pajak dan utang B adalah : " + str(hasil_b))
hasil_c = (pajak * laba_c) + (utang_c * 420000000)
print("maka total pajak dan utang C adalah : " + str(hasil_c))
hasil_d = (pajak * laba_d) + (utang_d * 550000000)
print("maka total pajak dan utang D adalah : " + str(hasil_d))
hasil_e = (pajak * laba_e) + (utang_e * 340000000)
print("maka total pajak dan utang E adalah : " + str(hasil_e))

print("")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
perusahaan = ['A-Xis', 'B-Log', 'C-in', 'D-Namic', 'E-psilon']
nilai = [23375000, 13330000, 51660000, 62045000, 11010000]
c = ['red', 'green', 'blue', 'red', 'green', 'blue']
plt.title("data pajak dan utang perusahaan")
plt.bar(perusahaan, nilai, color = c)
plt.show()

print("")
fig =plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
plt.title("DATA SISWA PPD SMK AL-MADANI GARUT")
perusahaan = ['A-Xis', 'B-Log', 'C-in', 'D-Namic', 'E-psilon']
nilai = [23375000, 13330000, 51660000, 62045000, 11010000]
c = ['red', 'green', 'blue', 'red', 'green', 'blue']
ax.pie(nilai,labels = perusahaan,autopct ='%1.2f%%')
plt.show()
```

```python
print("======== SOAL NO 5 ========")
tahun = 12

csr_1 = 0.025
csr_2 = 0.035

laba_a = 32500000
laba_b = 31000000
laba_c = 42000000
laba_d = 51500000
laba_e = 27000000

aset_a = 280000000
aset_b = 310000000
aset_c = 420000000
aset_d = 550000000
aset_e = 340000000


hasil_a = (laba_a * tahun) - csr_1 * aset_a - (7500000 * tahun) - 23375000
print("total pendapatan bersih pertahun perusahaan A : " + str(hasil_a))
hasil_b = (laba_b * tahun) - csr_2 * aset_b - (6090000 * tahun) - 13330000
print("total pendapatan bersih pertahun perusahaan B : " + str(hasil_b))
hasil_c = (laba_c * tahun) - csr_2 * aset_c - (15400000 * tahun) - 51660000
print("total pendapatan bersih pertahun perusahaan C : " + str(hasil_c))
hasil_d = (laba_d * tahun) - csr_2 * aset_d - (27510000 * tahun) - 62045000
print("total pendapatan bersih pertahun perusahaan D : " + str(hasil_d))
hasil_e = (laba_e * tahun) - csr_2 * aset_e - (8550000 * tahun) - 11010000
print("total pendapatan bersih pertahun perusahaan E : " + str(hasil_e))

print("")
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
perusahaan = ['A-Xis', 'B-Log', 'C-in', 'D-Namic', 'E-psilon']
nilai = [269625000, 274740000, 252840000, 206585000, 198490000]
c = ['red', 'green', 'blue', 'red', 'green', 'blue']
plt.title("total pendapatan bersih pertahun perusahaan")
plt.bar(perusahaan, nilai, color = c)
plt.show()

print("")
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
plt.title("total pendapatan bersih pertahun perusahaan")
perusahaan = ['A-Xis', 'B-Log', 'C-in', 'D-Namic', 'E-psilon']
nilai = [269625000, 274740000, 252840000, 206585000, 198490000]
c = ['red', 'green', 'blue', 'red', 'green', 'blue']
ax.pie(nilai,labels = perusahaan,autopct ='%1.2f%%')
plt.show()
```
